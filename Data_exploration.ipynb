```python
def delete_columns(dataset, rate = 0.8): 
    columns_to_drop = []
    for col in dataset.columns:
        print(f'{col} : {dataset[col].isna().sum()}')
        print(f'Filled percentage : {((dataset.shape[0] - dataset[col].isna().sum()) / dataset.shape[0])*100}')
        print('*'*10)
        
        if dataset[col].isna().sum() / len(dataset) > rate:
            columns_to_drop.append(col)
            
    print(f'Columns to drop : {columns_to_drop}')
    dataset = dataset.drop(columns = columns_to_drop)
    return dataset
```

```python
df = delete_columns(dataset)
```
